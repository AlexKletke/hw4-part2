### **Вывод:** ###
Цена зависит от 5 параметров:

- от количества ванных

- жилой площади

- оценки вида

- категории дома (grade)

- площади дома отдельно от фундамента (sqft_above)

```python
df['bathrooms'].nunique()
```

```python
data = df.groupby('bathrooms').agg({'price':'mean'})
data
```

```python
plt.figure(figsize=(8, 6))
plt.bar(data.index, data['price'])
plt.title('Взаимосвязь между ценой и количеством ванн')
plt.xlabel('Кол-во ванн')
plt.ylabel('Цена объекта с таким кол-вом ванн, млн');
```

## **ВЫВОД1:** ##
зависимость цены дома от количества ванн - средняя. Коэффициент корреляции = 0,5. В целом цена растет, но падение с ваннами в 7,5 показывает, что еще есть факторы.

```python
x = df['sqft_living']
y = df['price']
sum_x_y = 0
for i in range(0, len(x)):
    sum_x_y = sum_x_y + x[i]*y[i]
sum_y = 0
for number in y:
    sum_y += number
sum_x = 0
for number in x:
    sum_x += number
sum_x_2 = 0
for number in x:
    sum_x_2 += number**2
b1 = ((len(x) * sum_x_y - sum_x * sum_y)) / (len(x) * sum_x_2 - sum_x**2)
b0 = np.mean(y) - b1*np.mean(x)
```

```python
df['sqft_living'].nunique()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['sqft_living'], df['price'])
plt.plot(x, (b0 + b1 * x), 'r:' , label='средняя цена дома')
plt.title('Взаимосвязь между ценой и Жилой ПЛОЩАДЬЮ')
plt.ylabel('Цена в миллионах')
plt.xlabel('Кол-во жилой площади')
plt.legend()
plt.show()
```

## **ВЫВОД2:** ##
зависимость цены дома от количества жилой площади - высокая. Коэффициент корреляции = 0,7. Присутствует линейная зависимость.
