## **1.4 Изучите распределение года постройки** ##

```python
df['yr_built'].describe()
```

```python
df['yr_built'].unique()
```

```python
df['yr_built'].nunique()
```

```python
data = df.groupby('yr_built')['price'].agg(count='count').reset_index().sort_values(by='yr_built')
data
```

```python
plt.figure(figsize=(6, 4))

plt.plot(data['yr_built'], data['count'])

plt.title('Динамика постройки недвижимости по годам')
plt.xlabel('год постройки')
plt.ylabel('Кол-во построенных объектов');
```

## **ВЫВОД** ##
Выборка объектов недвижимости = 21613 за 115 лет. 

mean      1971

std         29 объектов

min       1900

25% построено до 1951 (первая четверть всех объектов построена за первые 51 год)

50% построено до 1975 (вторая четверть всех объектов построена за 24 года)

75%       1997 (третья четверть всех объектов построена за 22 года)

max       2015 (последняя четверть всех объектов построена за 18 лет)

Явный тренд на ускорение строящихся объектов: первые 50% было построено за 75 лет, а вторые 50% - за 40 лет. Причем темпы строительства возрастают с годами.
